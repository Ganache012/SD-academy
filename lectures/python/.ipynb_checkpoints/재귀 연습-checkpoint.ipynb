````python
def func(n):
    print("{}번 함수가 실행 되었습니다.".format(n))
    if n > 1: #기저조건 (Base Condition)
        print("Call Function")
    else:
        return print("재귀가 종료되었습니다.")
    func(n-1)
    return print("함수를 종료합니다.")

func(10)
````

````python
def func(n):
    print("{}번 함수가 실행 되었습니다.".format(n))
    if n <= 1: #기저조건 (Base Condition)
        print("기저 조건을 만족하였습니다.")
        return print("{}번 함수에서 재귀가 종료 되었습니다.".format(n))
    
    func(n-1)
    #return 은 함수의 종료를 의미적으로 나타내기 위해서 사용되었습니다.
    #여기서 return이 없이도 더 이상 실행할 코드가 없기 때문에
    # 함수는 종료 됩니다.
    return print("{}번 함수가 종료됩니다.".format(n))

func(5)
````

- 상향식 접근방법(동적 계획법), 하향식 접근 방법
- 상향식 접근방법: loop, 꼬리재귀(Tail Recursive)
- 하향식 접근방법: 재귀

# 상향식 접근 방법(loop)

ps) 같다(==), 같지않다(!=), 크다(>), 크거나 같다(>=), 작다(<), 작거나 같다(<=)

````python
n = 10
total = 0
while n != 0:
    total += n
    n -= 1
print(total)
````

# 상향식 접근 방법(꼬리재귀)

````python
def tail_recursive(total, n):
    print('{}번째 함수 호출의 total: {}'.format(n, total))
    # 기저 조건
    if n == 0:
        print("기저조건을 만족합니다. total: {}".format(total))
        return total
    
    ret = tail_recursive(total + n, n-1)
    print('{}번째 함수의 리턴: {}'.format(n, ret))
    return ret
    
tail_recursive(0, 10)
````

# 하향식 접근 방법(재귀)

````python
def recursive(n):
    # base condition
    if n != 0:
        return n + recursive(n-1)
    else:
        return n
print(recursive(10))
````

````python
def recursive(n):
    # base condition
    if n != 0:
        return n + recursive(n-1)
    else:
        return n
print(recursive(0))
````

````python
def recursive(n):
    print("Call recursive({})".format(n))
    # base condition
    if n != 0:
        ret = recursive(n-1)
        print("return {} + {}".format(n, ret))
        return n + ret
    else:
        print("return {}".format(n))
        return n

print(recursive(10))
````

## n**m을 구하는 프로그램 작성(꼬리재귀)

````python
def tail_recursive(n, m, mul):
    print('call multiple({}, {}, {})'.format(n, m-1, mul * n))
    if m >= 1:
        ret = tail_recursive(n, m-1, mul*n)
        print('return {}'.format(ret))
        return ret
    else:
        print('return {}'.format(mul))
        return mul
    
tail_recursive(2, 5, 1)
````

##  n**m을 구하는 프로그램 작성 (재귀)

### 내 code

````python
def multiple(n, m):
    print("Call multiple({})".format(m))
    if m >= 1:
        m = multiple(n, m-1)
        print("return {} * {}".format(n,m))
        return n * m
    else:
        print("return {}".format(m))
        return 1

multiple(3, 3)
````

### 강사님 code

````python
def multiple(n, m):
    print('call multiple({}, {})'.format(n,m))
    if m >= 1:
        ret = multiple(n, m-1)
        print('return {} * {}'.format(n,ret))
        return n * ret
    else: 
        print('return {}'.format(1))
        return 1

multiple(2, 5)
````

## 꼬리재귀

````python
# 꼬리재귀를 작성할 때에는
# 다음에 호출될 나 자신(Callee)에게 어떤 값을 전달할 것인지를 고민해봅시다.
# 이 문제에서는 각 단계별로 왼쪽 수가 홀수인 경우 오른쪽수를 더한
# total을 다음의 나 자신에게 넘겨주는 것이 핵심이다.
def farmer_recursive(n, m, total):
    if n >= 1:
        if (n % 2) == 1:# 홀수인 경우
            return farmer_recursive(n//2, m*2, total+m)
        else:
            return farmer_recursive(n//2, m*2, total)
    else:
        return total
    
print(farmer_recursive(25, 13, 0))
````

## 재귀

````python
# 재귀를 작성 할 때에는 마지막 함수가 호출된 이후로
# 어떤값을 이전에 호출했던 함수(Caller)에게 돌려줄지(return)를 고민해본다.
def russian_recursive(n, m):
    if n >= 1:
        if (n % 2) == 1: #홀수인 경우
            return m + russian_recursive(n//2, m*2)
        else: 
            return russian_recursive(n//2, m*2)
    else:
        return 0
    
russian_recursive(25, 13)
````

# 팩토리얼 구하기
```
   n! = 1 * 2 * 3 * .... * n
   5! = 5 * 4 * 3 * 2 * 1
```
- 1부터 n까지의 합을 구하는 내용과 동일

# 루프(while)를 이용한 팩토리얼 구하기!

````python
n = int(input())
result = 1 #항등원
while 1 <= n:
    result *= n
    n -= 1
    print("result: {}, n: {}".format(result, n))
print(result)
````

# 꼬리재귀를 이용한 팩토리얼 구하기!

````python
def factorial_recursive(n, result):
    print('n: {}, result: {}'.format(n, result))
    if n >= 1:
        ret = factorial_recursive(n-1, result*n)
        print('result: {}'.format(ret))
        return ret
    else:
        print('result: {}'.format(result))
        return result
    
factorial_recursive(10, 1)
````

# 재귀를 이용한 팩토리얼 구하기!

````python
def factorial(n):
    print('call n: {}'.format(n))
    if n >= 1:
        ref = n*factorial(n-1)
        print('n:{}, ref:{}'.format(n,ref))
        return ref
    else:
        print('return: {}'.format(1))
        return 1
    
factorial(10)
````

# 피보나치 수 구하기 (while loop)

### 리스트를 이용한 피보나치 수 구하기

```
    list_ = [1, 2] #0번째 피보나치 수와 1번째 피보나치수는 정해져있음(점화식)
    list_[2] = list_[0] + list_[1]
    list_[3] = list_[1] + list_[2]
    ....
    list_[n] = list_[n-1] + list_[n-2]
    
```

### 리스트를 사용하지 않는 경우 피보나치 수 구하기

```
    loop: 0
        prev_no = 0
        current_no = 1
    loop: 1
        current_no(1) = current_no(0) + prev_no(0)
        prev_no(1), current_no(1) = current_no(1), prev_no(0)
    loop: 2
        current_no(2) = current_no(1)=0 +prev_no(1)=1
        prev_no(2), current_no(2) = current_no(2), prev_no(1)
    loop: 3
        current_no(3) = current_no(2) + prev_no(2)
        prev_no(3), current_no(3) = current_no(3), prev_no(2)
    loop: 4
        current_no(4) = current_no(3) + prev_no(3)
        prev_no(4) current_no(4) = current_no(4), prev_no(3)
    
    .....
    
    [0, 1, 1, 2, 3...]
```

````python
#점화식
#f0 = 0
#f1 = 1

prev_no = 0
current_no = 1

n= int(input())

i =0
while i <= n:
    current_no = prev_no + current_no
    print('{} -> 피보나치 수: {}'.format(i, current_no))
    prev_no, current_no = current_no, prev_no
    print('{} -> prev no: {}, current no: {}'.format(i, prev_no, current_no))
    
    i += 1

print(current_no)  
````

# 피보나치 수 구하기(꼬리재귀)

### 내 code

````python
# next: prev_no, before: current_no
def fibonacci_tail1(n, prev_no, current_no):
    print('n: {}, prev_no: {}, current_no: {}'.format(n, prev_no, current_no))
    if n == 0:
        print('prev_no: {}'.format(current_no))
        return current_no
    else:
        ref = fibonacci_tail1(n-1, current_no, prev_no + current_no)
        print('n:{}, ref:{}'.format(n,ref))
        return ref

fibonacci_tail1(5, 1, 0)
````

````python
#next: current_no, before: prev_no
def fibonacci_recursive(n, prev_no, current_no):
    print('n: {}, prev_no: {}, current_no: {}'.format(n, prev_no, current_no))
    if n > 0 :
        re = fibonacci_recursive(n-1, current_no, prev_no + current_no)
        print('n: {}, re:{}'.format(n, re))
        return re
    else:
        return prev_no

fibonacci_recursive(10, 0, 1)
````

### 강사님 code

````python
#기존 피보나치 수열은 current =1, prev=0인데 쉬운 구현을 위해 두 변수의 값을 swith함
def fibo(n, current = 0, prev = 1):
    if n > 0:
        return fibo(n-1, current+prev, current)
    return current
print(fibo(10))
````

# 피보나치 수 구하기(재귀)

### 내 code

````python
def fibonacci_tail(n):
    print('Call:{}'.format(n))
    if n == 0:
        print('return: {}'.format(0))
        return 0
    elif n == 1:
        print('return: {}'.format(1))
        return 1
    else:
        ret = fibonacci_tail(n-1)+fibonacci_tail(n-2)
        print('n: {}, ret:{}'.format(n, ret))
        return ret
    
fibonacci_tail(5)
````

### 강사님 code

````python
#트리 재귀
n = int(input())

def fibo(n):
    if n<=1:
        return n
    return fibo(n-1) + fibo(n-2)

print(fibo(n))
````

### 별찍기 -10단계 _ 쪼개기

````python
def board(i, j, n):
    if n == 1: #기저 조건
        # 준비된 판에 색을 칠한다.
        independent_door[i][j] = "*"
        return # 어떤 무언가를 돌려주는게 아니라,그냥 함수를 종료     
    # n = 3이면 1, n=9이면 3, n=27이면 9 
    add = n//3
    
    # 일단 처음에는 반복문 없이 작성
    board(i, j, add)
    board(i, j+add, add)
    board(i, j+add+add, add)
    
    board(i+add, j, add)
    #board(i+add, j+add, add)
    board(i+add, j+add+add, add)
    
    board(i+add+add, j, add)
    board(i+add+add, j+add, add)
    board(i+add+add, j+add+add, add)

n = 81
independent_door = [[' ' for _ in range(n)] for _ in range(n)]

board(0,0,81)
for i in independent_door:
    print(''.join(i))
````

### 내가 고민해본 code

````python
def board(a, b, n):
    if n == 1: #기저 조건
        # 준비된 판에 색을 칠한다.
        independent_door[a][b] = "*"
        return # 어떤 무언가를 돌려주는게 아니라,그냥 함수를 종료     
    # n = 3이면 1, n=9이면 3, n=27이면 9 
    add = n//3
    
    # 반복문
    for i in range(3):
        for j in range(3):
            if (i==1 and j==1):
                pass
            else:
                board(a+(i*add),b+(j*add),add)
        

n = 27
independent_door = [[' ' for _ in range(n)] for _ in range(n)]

board(0,0,27)

for i in independent_door:
    print(''.join(i))
````

### 강사님 code _재귀

````python
def board(i, j, n):
    if n == 1: #기저 조건
        # 준비된 판에 색을 칠한다.
        independent_door[i][j] = "*"
        return # 어떤 무언가를 돌려주는게 아니라,그냥 함수를 종료     
    # n = 3이면 1, n=9이면 3, n=27이면 9 
    add = n//3
    
    for p in range(3):
        for q in range(3):
            if p == 1 and q ==1 : continue
            board(i+add*p, j+add*q, add)

n = int(input())
independent_door = [[' ' for _ in range(n)] for _ in range(n)]

board(0,0,n)
for i in independent_door:
    print(''.join(i))
````

# 합병 정렬(Merge Sort)
- 분할-정복 기법을 사용하는 대표적인 알고리즘
- 안정적으로 빠른 시간내에 동작하는 정렬 알고리즘
- 재귀 호출도 연습해볼 수 있다.

````python
import random

unsort = [x for x in range(7)]
random.shuffle(unsort)
unsort
````

````python
# 정렬하고 합치기
def merge(left, right):
    print('left: {}, right:{}'.format(left, right))
    sort = []
    while left != [] and right !=[]: #left, right둘중 하나의 리스트가 비워지면 루프종료
        if left[0] < right[0]:
            sort.append(left.pop(0))
        else:
            sort.append(right.pop(0))
    if len(left): sort += left
    else: sort += right
        
    print('sort:{}'.format(sort))
    return sort#left, right의 정렬된 리스트 
# 분할
# array == list
def divide(arr):
    print(arr)
    if len(arr) == 1: return arr #리스트내의 원소가 하나만 남을 때까지
    
    mid = len(arr) // 2
    return merge(divide(arr[:mid]), divide(arr[mid:]))

divide(unsort)
````

````python
# 정렬하고 합치기
def merge(left, right):
    result = []
    while len(left)> 0 or len(right)>0:
        print('len left: '.format(len(left)), 'left list: '.format(left))
        print('len right: '.format(len(right)), 'right list: '.format(right))
        #둘다 0보다 크다
        if len(left) >0 and len(right)>0:
            if left[0] <= right[0]:
                result.append(left[0])
                left = left[1: ]
            else:
                result.append(right[0])
                right = right[1: ]
        #둘다 0보다 크진 않지만, left가 0보다 크면
        elif len(left)>0:
            result.append(left[0])
            left = left[1: ]
        elif len(right)>0:
            result.append(right[0])
            right = right[1: ]
        print('result: '.format(result))
        return result
````
