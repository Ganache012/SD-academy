- 차분
    - 예를 들면 입력변수 x중에, 마케팅 비용
    - 반응변수 y를 판매율 
    - 오늘 지출한 마케팅 비용에 대해서 꼭 오늘 판매에만 영향을 끼친다고 볼 수 없겠지?
    - 지난주에 지출한 마케팅 비용에 대해서 오늘 효과가 나타날 수도 있다..
    - 변수들의 값을 한칸씩 미루는 거다. 

```python
rawData['cnt'].shift(1)
```

```python
rawData['cnt'].shift(2)
```
