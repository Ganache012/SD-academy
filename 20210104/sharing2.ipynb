```python
boston_ridge_fit = Ridge(alpha=0.5, fit_intercept=False, normalize=True, random_state=1234).fit(x_train, y_train)
```

```python
pd.DataFrame(boston_ridge_fit.coef_, columns=['alpha'])
```

```python
# alpha 회귀계수 살펴보기
# alpha 후보 뽑기
result_ridge = pd.DataFrame()
alphas = np.logspace(-2, 1, 4)
for alpha in alphas:
    boston_ridge_fit = Ridge(alpha=alpha, fit_intercept=False, normalize=True, random_state=1234).fit(x_train, y_train)
    result_coef = pd.DataFrame(boston_ridge_fit.coef_, columns=['alpha={}'.format(alpha)])
    result_ridge = pd.concat([result_ridge, result_coef], axis=1)
```

```python
result_ridge
```

```python
result_ridge.T.plot(figsize=(12,7), legend=False)
plt.grid()
plt.show()
```

```python
alphas = np.logspace(-2, 2, 10)
for alpha in alphas:
    bostion_ridge_fit = Ridge(alpha = alpha, fit_intercept=False, normalize=True, random_state=1234).fit(x_train, y_train)
    print('score: {}'.format(boston_ridge_fit.score(x_train, y_train)))
```

```python
# next-time(2020-12-30)
alphas = np.logspace(-2, 1, 10)
parameters = {'alpha': alphas}

ridge = Ridge()
result_ridge = GridSearchCV(ridge, parameters, scoring='neg_mean_squared_error', cv=5)
result_ridge.fit(x_train, y_train)
result_ridge.best_params_
```

1. numpy

    - R에서 지원하는 라이브러리를 통해서 실습

3. Python
    - 파이썬에서는 따로 선형회귀, 비선형회귀, 다항회귀, 로지스틱 회귀 등에 대한 라이브러리는 따로 확인을 하지 않았다.

## 단순선형회귀
- 자료: 보스턴 집값데이터 사용

- 동전을 던져서 앞면이 나오면 앞으로 한걸음, 뒷면이 나오면 뒤로 한걸음 간다고 했을 때,
    - 앞면이 나오는 횟수와

### 단순선형회귀 분석
- 입력변수 x와 반응변수 y 사이에 선형적인 관계를 가정한다는 뜻
- 반응변수 y는 레이블이 된다.
- 입력변수 x는 단일변수가 된다.(ex 범죄율과 집값의 관계, 하위계층의 비율, 방의 갯수, ..)
    - 예를 들면 야구선수(용병)들 보면 국내의 치안환경을 굉장히 높게 평가)

# VIF를 통한 다중공선성 제거
- 종속변수(출력변수,...)Y는 입력변수 X들에 대해 종속적이다.
- 입력변수 X들은 각각에 대하여 서로 독립임을 가정
    - 즉 Y를 제외하고는 각 변수들은 다른 변수들에 영향을 주거나 받아선 안된다.
- 모형이 제대로 적합하지 않거나, 아예 추정이 불가능한 경우도 발생
- 종속변수 Y를 제외한 X들에 대해서
- 특정변수 x를 종속변수로 해서 나머지 x(n-1)들에 대해 선형회귀를 적합 하는 방법
    - x와 x들에 대한 관계를 확인하는 방법
    - 모형의 추정치가 낮으면 해당 변수는 다른 변수들에 대해 그만큼 영향력이 없음을 뜻한다.
    - 그러나 반대로 추정치가 높으면 영향력이 많음을 뜻한다.

## VIF를 통한 다중공선성 확인
