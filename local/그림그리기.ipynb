# matplotlib를 이용한 그래프 표현하기

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 더미 데이터 생성

```python
x = np.linspace(0, 10, 11)
y = x ** 2 + x + 2 + np.random.randn(11)

print(x)
print(y)
```

## 그래프 그리기
- plot함수(선, 그래프), scatter(점), hist(히스토그램) 등을 사용
    - 기본적으로 x,y에 해당하는 값이 필요

```python
plt.plot(x, y)
```

```python
plt.scatter(x,y)
```

## 그래프에 주석 추가

```python
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('title')
plt.grid()
plt.plot(x,y)
```

```python
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('text')
plt.grid()

plt.xlim(-20, 20)
plt.ylim(-200, 200)

plt.plot(x,y)
```

## 한 화면에 여러 그래프 출력하기
- subplot

```python
plt.figure(figsize=(20, 10))

plt.subplot(2,2, 1)
plt.plot(x, y, 'r')

plt.subplot(2, 2, 2)
plt.plot(y, x, 'g')

plt.subplot(2,2,3)
plt.plot(x, y, 'b')

plt.subplot(2,2,4)
plt.plot(y, x, '#000000')
```

# matplotlib의 또 다른 그림들

## 막대그래프
- 바차트

```python
x = np.arange(5)
xlabel = ['첫번째', '두번째', '세번째', '네번째', '다섯번째']
y = [100, 200, 300, 400, 500]

plt.bar(x, y)
plt.xticks(x, xlabel)
plt.show()
```

```python
#bar chart 디자인
x = np.arange(5)
y = [100, 200, 300, 400, 500]

plt.bar(x, y, width=0.5, align='center', color='#123456', edgecolor='y', linewidth=3)
plt.title('안녕?')
plt.xlabel('x축')
plt.show()
```

```python
#bar chart 디자인
#뒤집기
x = np.arange(5)
y = [100, 200, 300, 400, 500]

plt.barh(x, y, height=0.5, align='center', color='#123456', edgecolor='y', linewidth=3, alpha=0.8)
plt.title('안녕?')
plt.xlabel('x축')
plt.show()
```

## 히스토그램
- y가 필요 없기 때문에, 단변량 그림 그리기에 좋다.
- bins로 히스토그램의 바의 갯수를 설정

```python
# 히스토그램
data = np.random.randint(1, 100, size=200)
print(data)
```

```python
plt.hist(data, bins=30, alpha=0.3)
plt.grid()
plt.show()
```

## Scatter(산점도)

```python
N = 100
x = np.random.randn(N)
y = np.random.randn(N)
area = (20 * np.random.rand(N))**2
color = np.sqrt(area)
plt.scatter(x,y, s=area, c=color, alpha=0.8)
plt.show()
```
