```python
import warnings
warnings.filterwarnings('always')
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 통계적 모형
import statsmodels.api as sm

# 머신러닝 관련
from sklearn import datasets
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.model_selection import train_test_split, GridSearchCV #GridSearchCV : AutoML
# 머신러닝 모델 평가 관련
from sklearn.metrics import mean_squared_log_error, mean_squared_error,\
r2_score, mean_absolute_error
```

```python
rawData = pd.read_csv('./data/ToyotaCorolla.csv')
```

```python
rawData.info()
```

```python
rawData.describe()
```

```python
rawData
```

```python
rawData.drop(['Model', 'Fuel_Type'], axis='columns', inplace=True)
```

```python
rawData
```

```python
rawData.set_index('Id', inplace=True)
```

```python
rawTrain = rawData.loc[rawData.index < 1148, :]
rawTest = rawData.loc[rawData.index >= 1148, :]
print(rawTrain.shape, rawTest.shape)
```

```python
rawData.dropna()
```

### No-Random sampling

```python
x_train = rawTrain.loc[:, 'Age_08_04':]
y_train = rawTrain[['Price']]
x_test = rawTest.loc[:, 'Age_08_04':]
y_test = rawTest[['Price']]
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
model1 = sm.OLS(y_train, x_train).fit()
model1.summary()
```

```python
predict_train = model1.predict(x_train).values
predict_test = model1.predict(x_test).values
```

```python
predict_train
```

```python
y_train
```

```python
pd.concat([y_train, pd.DataFrame(predict_train, index=y_train.index,\
                                 columns=['predict'])], axis=1).plot(kind='line', figsize=(21,7))
plt.show()
```

```python
np.mean((y_train.values.flatten() - predict_train) ** 2)
```

```python
pd.concat([y_test, pd.DataFrame(predict_test, index=y_test.index,\
                                 columns=['predict'])], axis=1).plot(kind='line', figsize=(21,7))
plt.show()
```

```python
np.mean((y_test.values.flatten() - predict_test) ** 2)
```

### 랜덤샘플링

```python
x = rawData.loc[:, 'Age_08_04':]
y = rawData['Price']

train_x, test_x, train_y, test_y = train_test_split(x,y, train_size=0.8, test_size=0.2, random_state=1234)
```

```python
model1 = sm.OLS(train_y, train_x).fit()
model1.summary()
```

```python
predict_train_r = model1.predict(train_x).values
predict_test_r = model1.predict(test_x).values
```

```python
predict_train_r
```

```python
train_y
```

```python
pd.concat([train_y, pd.DataFrame(predict_train_r, index=train_y.index,\
                                 columns=['predict'])], axis=1).plot(kind='line', figsize=(21,7))
plt.show()
```

```python
pd.concat([test_y, pd.DataFrame(predict_test_r, index=test_y.index,\
                                 columns=['predict'])], axis=1).plot(kind='line', figsize=(21,7))
plt.show()
```
