## 선형적인 관계로 가정(모수적 추정)
- y = ax + b
- 주어진 데이터를 가장 잘 설명할 수 있는 선 하나를 찾는 것

## error function(오차함수) or cost function

- Squared Error
- Mean Squared Error(MSE)

$$
    \sum_{i=1}^n | Y_i - (ax_i + b) |^2
$$

1. 주어진 선에서: ax + b
2. 어떤 기준으로: squared error
3. 가장 좋은: 최소로 하는(오차를)
4. 선을 찾는 것입니다: a, b

$$
    (a,b) = argmin_{a,b}\sum_{i=1}^n | Y_i - (ax_i + b) |^2
$$

## 2차 함수의 최소값을 구하기

$$
    f(x) = x^2 - 4x + 6
$$

```python
def f(x):
    return x ** 2 - 4 * x + 6
```

```python
x = np.linspace(-5, 5, 101)
print(x)
```

```python
y = f(x) # y를 구한다
```

```python
# 함수의 최솟값은 2가 된다
plt.plot(x,y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 최소값 구하기
xidx = np.argmin(y)
argx = x[xidx]
valueF = f(argx)
print('argX:',x[xidx])
print('f value:',valueF)
```

```python
plt.plot(x,y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## 최급하강법(steepest descent) or 경사하강법(Gradient descent)

$$
    x_{k+1} = x_k - \alpha \nabla f(x_k) \\
    x^{(k+1)} = x^{(k)} -  \alpha \nabla f(x^{(k)})
$$

```python
def grad_fx(x):
    return 2 * x + 4
```

```python
x0 = 0
learning_rate = 0.01

x1 = x0 - learning_rate * grad_fx(x0)
print('argx:{}, slope: {}, f value: {}'.format(x1, grad_fx(x1), f(x1)))
```

```python
x0 = x1
x1 = x0 - learning_rate * grad_fx(x0)
print('argx:{}, slope: {}, f value: {}'.format(x1, grad_fx(x1), f(x1)))
```

```python
x0 = x1
x1 = x0 - learning_rate * grad_fx(x0)
print('argx:{}, slope: {}, f value: {}'.format(x1, grad_fx(x1), f(x1)))
```

```python
x0 = x1
x1 = x0 - learning_rate * grad_fx(x0)
print('argx:{}, slope: {}, f value: {}'.format(x1, grad_fx(x1), f(x1)))
```

```python
# 한번에 근접해지는 것을 보기
x0 = 0
learning_rate = 0.01
maxIter = 100

for it in range(maxIter):
    x1 = x0 - learning_rate * grad_fx(x0)
    print('argx:{}, slope: {}, f value: {}'.format(x1, grad_fx(x1), f(x1)))
    x0=x1
```

## 이변수일 경우

$$
    f(x, y) = (x-2)^2 + (y-2)^2
$$

### 미분계수 구하기

```python
x = sp.symbols('x')
y = sp.symbols('y')
f = (x-2)**2 + (y-2) **2
print(sp.diff(f,x))
print(sp.diff(f,y))
```

```python
def f(x, y):
    return (x-2) ** 2 + (y-2) ** 2

def grad_fx(x,y):
    return 2 * x - 4

def grad_fy(x,y):
    return 2 * y - 4
```

```python
x, y = np.meshgrid(np.linspace(-5, 5, 101), np.linspace(-5, 5, 101))
z = f(x,y)
```

```python
np.argmin(z)
```

```python
x0 = np.array([-5, -5])
learning_rate = 0.01
maxIter = 1000
#grad = np.array([grad_fx(x0[0], x0[1]), grad_fy(x0[0], x0[y])]) # 편미분값의 벡터 표현

for it in range(maxIter):
    grad = np.array([grad_fx(*x0), grad_fy(*x0)]) # 인자가 여러개 들어가는 경우 사용하는 거
    x1 = x0 - learning_rate * grad
    print('argx:{}, slope: {}, f value: {}'.format(x0, grad, f(*x0)))
    x0 = x1
```

## 변수가 3 개 일때

$$
    f(x, y, z) = x^2z^2 + 4x + 6 + y^2 +xy
$$

- 함수의 값을 최소로 하는 x,y,z를 구해봅시다.

```python
f = lambda x,y,z: x ** 2  * z ** 2 + 4 * x + 6 + y ** 2 + x*y
grad_fx = lambda x,y,z: 2 * x * z ** 2 + y + 4
grad_fy = lambda x,y,z: x + 2 * y
grad_fz = lambda x,y,z: 2 * x ** 2 * z
```

```python
x,y,z = np.meshgrid(np.linspace(-5, 5, 101), np.linspace(-5,5,101), np.linspace(-5,5,101))
k = f(x,y,z)
```

```python
x0 = np.array([0, 0, 0])
learning_rate = 0.1
maxIter = 100

for it in range(maxIter):
    grad = np.array([grad_fx(*x0), grad_fy(*x0), grad_fz(*x0)])
    x1 = x0 - learning_rate * grad 
    print('argx: {}, slope:{}, f value: {}'.format(x0, grad, f(*x0)))
    x0 = x1
```
