## 주어진 문자열을 이용해 만들 수 있는 모든 문자열의 조합

- 어떻게 만들어 낼 수 있을 것인가?

````python
#완전탐색의 예: 순서와 중복에 상관없이 모든 조합을 찾는 방법
alphabet = ['a', 'b', 'c']

for first in alphabet:
    for second in alphabet:
        for third in alphabet:
            print('{}{}{}'.format(first, second, third))
````

````python
#완전탐색의 예: 중복되지 않은 조합일 경우(중복만 고려, 순서 상관없음)
alphabet = 'abc'

for i in range(0, len(alphabet)):
    for j in range(i+1, len(alphabet)):
        for k in range(j+1, len(alphabet)):
            print('{}{}{}'.format(alphabet[i], alphabet[j], alphabet[k]))
````

````python
#완전탐색의 예: 중복되지 않은 조합일 경우(중복만 고려, 순서 상관없음)
#alphabet = 'abc' #튜플(비가변형)이기 때문에 제거가 되지 않음
alphabet = ['a', 'b', 'c']
memoization1 = []

for i in range(0, len(alphabet)):
    # 선택된 원소를 제외하고 다음 루프에 전달(쉬움)
    if len(memoization1) != 0:
        alphabet.insert(i-1, memoization1.pop())
    memoization1.append(alphabet.pop(i))
    print(alphabet, memoization1)
````

````python
# 스택 또는 큐를 이용한 조합
#완전탐색의 예: 중복되지 않은 조합일 경우(중복만 고려, 순서 상관없음)
alphabet = ['a', 'b', 'c']
memoization = []

for i in range(0, len(alphabet)):
    # 선택된 원소를 제외하고 다음 루프에 전달(쉬움)
    if len(memoization) != 0:
        alphabet.insert(i-1, memoization.pop())
    
    #print(alphabet[i],end="")
    memoization.append(alphabet.pop(i))
    print(alphabet, memoization)
    for j in range(len(alphabet)):
        if len(memoization) != 0:
            alphabet.insert(j-1, memoization.pop())
            
        print(alphabet[j], end="")
        memoization.append(alphabet.pop(j))
        
        for k in range(len(alphabet)):
            print(alphabet[k])
````

loop 0:

````python
#완전탐색의 예: 중복되지 않은 조합일 경우(중복만 고려, 순서 상관없음)
alphabet = ['a', 'b', 'c']
memoization = []

for i in range(0, len(alphabet)):
    # 선택된 원소를 제외하고 다음 루프에 전달(쉬움)
    if len(memoization) != 0:
        alphabet.insert(i-1, memoization.pop())
    
    print(alphabet[i],end="")
    memoization.append(alphabet.pop(i))
    #print(alphabet, memoization)
    for j in range(len(alphabet)):
        if len(memoization) != 1:
            alphabet.insert(j-1, memoization.pop(-1))
            
        print(alphabet[j], end="")
        memoization.append(alphabet.pop(j))
        
        for k in range(len(alphabet)):
            print(alphabet[k])
            
    alphabet.append(memoization.pop(-1))
````

## 알파벳 소문자에서 만들 수 있는 3글자 조합

### 강사님 code

````python
#완전탐색의 예: 중복되지 않은 조합일 경우(중복만 고려, 순서 상관없음)
# 따라서 루프가 3개기 때문에 더 많은 문자를 입력해도 3개씩 조합되어 나온다
alphabet = list('abcdefghijklmnopqrstuvwxyz')
memoization=[]

for i in range(0, len(alphabet)):
    # 선택된 원소를 제외하고 다음 루프에 전달(쉬움)
    if len(memoization) != 0:
        alphabet.insert(i-1, memoization.pop(-1))
    memoization.append(alphabet.pop(i))
    for j in range(len(alphabet)):
        if len(memoization) != 1:
            alphabet.insert(j-1, memoization.pop(-1))
        memoization.append(alphabet.pop(j))
        for k in range(len(alphabet)):
            print('{}{}'.format(''.join(memoization), alphabet[k]))
        alphabet.append(memoization.pop(-1))
````

## 알파벳 소문자에서 만들 수 있는 4자리 조합

````python
#완전탐색의 예: 중복되지 않은 조합일 경우(중복만 고려, 순서 상관없음)
# 따라서 루프가 3개기 때문에 더 많은 문자를 입력해도 3개씩 조합되어 나온다
alphabet = list('abcdefghijklmnopqrstuvwxyz')
memoization=[]

for i in range(0, len(alphabet)):
    # 선택된 원소를 제외하고 다음 루프에 전달(쉬움)
    if len(memoization) != 0:
        alphabet.insert(i-1, memoization.pop(-1))
    memoization.append(alphabet.pop(i))
    for j in range(len(alphabet)):
        if len(memoization) != 1:
            alphabet.insert(j-1, memoization.pop(-1))
        memoization.append(alphabet.pop(j))
        for k in range(len(alphabet)):
            if len(memoization) != 2:
                alphabet.insert(k-1, memoization.pop(-1))
            memoization.append(alphabet.pop(k))
            for l in range(len(alphabet)):
                print('{}{}'.format(''.join(memoization), alphabet[l]))
        alphabet.append(memoization.pop(-1))
````

### 중복과 순서를 모두 고려한 조합

````python
alpha = list('abcdef')

size = len(alpha)

for i in range(size-2):
    for j in range(i+1, size-1):
        for k in range(j+1, size):
            print('{}{}{}'.format(alpha[i], alpha[j], alpha[k], end=''))
    print()
````

````python
n, m = map(int, input().split())
a = list(map(int, input().split()))
b = len(a)
plus = 0
for i in range(0, b - 2):
        for j in range(i + 1, b - 1):
                for k in range(j + 1, b):
                        if a[i] + a[j] + a[k] > m:
                                continue
                        else:
                            plus = max(plus, a[i] + a[j] + a[k])
print(plus)
````

### 강사님 code

````python
n,m = input().split()
n, m = int(n), int(m)
card = input().split()

for i in range(n):
    card[i] = int(card[i])
    
num = 0
for i in range(n-2):
    for j in range(i+1, n-1):
        for k in range(j+1, n):
            tmp = card[i] + card[j] + card[k]
            if num < tmp <= m: num = tmp
print(num)
````
